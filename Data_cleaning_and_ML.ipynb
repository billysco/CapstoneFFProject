```python
import pandas as pd
import os
from pathlib import Path
import glob
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def removeprefix(input_str, prefix):
    if prefix and input_str.startswith(prefix):
        return input_str[-len(prefix):]
    return input_str
```

```python
def get_data(year):
    path = os.getcwd() + f"\\{year}"
    csv_files = glob.glob(os.path.join(path, "*.csv"))
    my_list = [x for x in csv_files]
    dfs = []
    for f in my_list:
        df = pd.read_csv(f)
        df['Week'] = f[-6:-4]
        dfs.append(df)
#         year+'week'+counter = df.copy()
    
    final_df = pd.concat(dfs)
    final_df.drop(columns=['StandardFantasyPoints', 'HalfPPRFantasyPoints'], axis=1, inplace=True)
    final_df['Year'] = year

    return final_df
    
get_data(1999)
```
